# COVID-19 Tweet Sentiment Classification with RNN Models

This notebook implements recurrent neural network models for text classification using the Kaggle dataset **Coronavirus tweets NLP - Text Classification** by Datatattle.

Dataset URL: https://www.kaggle.com/datasets/datatattle/covid-19-nlp-text-classification

The goal is to compare a custom RNN model trained from scratch against a transfer-learning model initialized with pre-trained GloVe word embeddings.

## Project Overview

This is a supervised five-class sentiment classification task. Each input is a COVID-19 related tweet, and the target label is one of:

- `Extremely Negative`
- `Negative`
- `Neutral`
- `Positive`
- `Extremely Positive`

Models compared:

1. **Custom architecture:** a bidirectional GRU or LSTM model with randomly initialized trainable embeddings.
2. **Transfer learning:** the same recurrent classifier initialized with pre-trained GloVe embeddings and fine-tuned on this dataset.

````python
import html
import random
import re
from collections import Counter
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import torch
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.preprocessing import LabelEncoder
from torch import nn
from torch.utils.data import DataLoader, TensorDataset, random_split

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
sns.set_theme(style="whitegrid")
print("Device:", DEVICE)
````

````python
DATA_DIR = Path("data/covid19-nlp-text-classification")
TRAIN_PATH = DATA_DIR / "Corona_NLP_train.csv"
TEST_PATH = DATA_DIR / "Corona_NLP_test.csv"

def require_file(path: Path) -> None:
    if not path.exists():
        raise FileNotFoundError(f"Missing {path}. Download the Kaggle dataset and place the file here.")

require_file(TRAIN_PATH)
require_file(TEST_PATH)

# The CSV files require latin-style encoding; UTF-8 can fail on this dataset.
train_df = pd.read_csv(TRAIN_PATH, encoding="ISO-8859-1")
test_df = pd.read_csv(TEST_PATH, encoding="ISO-8859-1")

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
display(train_df.head())
````

````python
summary = pd.DataFrame({
    "split": ["train", "test"],
    "instances": [len(train_df), len(test_df)],
    "columns": [train_df.shape[1], test_df.shape[1]],
    "missing_locations": [train_df["Location"].isna().sum(), test_df["Location"].isna().sum()],
})
display(summary)

class_distribution = pd.DataFrame({
    "train": train_df["Sentiment"].value_counts().sort_index(),
    "test": test_df["Sentiment"].value_counts().sort_index(),
}).fillna(0).astype(int)
display(class_distribution)

ax = class_distribution.plot(kind="bar", figsize=(10, 5), rot=30)
ax.set_title("Sentiment class distribution")
ax.set_xlabel("Sentiment")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
````

## Text Preprocessing

Applied preprocessing:

1. Decode HTML entities such as `&amp;`.
2. Remove URLs.
3. Remove Twitter mentions.
4. Preserve hashtag words while removing `#`.
5. Lowercase text.
6. Remove non-alphanumeric characters except apostrophes.
7. Collapse repeated whitespace.
8. Build a vocabulary from train text only.
9. Convert tweets to padded token ID sequences.

Labels are encoded with `LabelEncoder`.

````python
URL_RE = re.compile(r"https?://\S+|www\.\S+")
MENTION_RE = re.compile(r"@\w+")
HASHTAG_RE = re.compile(r"#(\w+)")
NON_TEXT_RE = re.compile(r"[^a-zA-Z0-9' ]+")
SPACE_RE = re.compile(r"\s+")

def clean_tweet(text: str) -> str:
    text = html.unescape(str(text))
    text = URL_RE.sub(" ", text)
    text = MENTION_RE.sub(" ", text)
    text = HASHTAG_RE.sub(r"\1", text)
    text = text.lower()
    text = NON_TEXT_RE.sub(" ", text)
    text = SPACE_RE.sub(" ", text).strip()
    return text

train_df = train_df.copy()
test_df = test_df.copy()
train_df["clean_text"] = train_df["OriginalTweet"].apply(clean_tweet)
test_df["clean_text"] = test_df["OriginalTweet"].apply(clean_tweet)
display(train_df[["OriginalTweet", "clean_text", "Sentiment"]].head())
````

````python
MAX_VOCAB_SIZE = 30000
MAX_SEQUENCE_LENGTH = 80
PAD_TOKEN = "<PAD>"
OOV_TOKEN = "<OOV>"

def tokenize(text: str) -> list[str]:
    return text.split()

def build_vocab(texts: pd.Series, max_vocab_size: int) -> dict[str, int]:
    counter = Counter()
    for text in texts:
        counter.update(tokenize(text))
    vocab = {PAD_TOKEN: 0, OOV_TOKEN: 1}
    for word, _ in counter.most_common(max_vocab_size - len(vocab)):
        vocab[word] = len(vocab)
    return vocab

def encode_text(text: str, vocab: dict[str, int], max_length: int) -> list[int]:
    ids = [vocab.get(token, vocab[OOV_TOKEN]) for token in tokenize(text)[:max_length]]
    ids += [vocab[PAD_TOKEN]] * (max_length - len(ids))
    return ids

vocab = build_vocab(train_df["clean_text"], MAX_VOCAB_SIZE)
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_df["Sentiment"])
y_test = label_encoder.transform(test_df["Sentiment"])
class_names = list(label_encoder.classes_)

X_train = np.array([encode_text(text, vocab, MAX_SEQUENCE_LENGTH) for text in train_df["clean_text"]], dtype=np.int64)
X_test = np.array([encode_text(text, vocab, MAX_SEQUENCE_LENGTH) for text in test_df["clean_text"]], dtype=np.int64)

num_classes = len(class_names)
vocab_size = len(vocab)
print("Classes:", class_names)
print("Vocabulary size:", vocab_size)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
````

## Performance Metrics

The primary metric is **macro F1-score**.

Justification: the dataset is moderately imbalanced and the minority classes are meaningful sentiment categories. Macro F1 gives equal weight to every class, while accuracy can hide poor performance on `Extremely Negative`, `Neutral`, or `Extremely Positive`.

Supporting metrics: accuracy, macro precision, macro recall, weighted F1, classification report, and confusion matrix.

````python
def make_loaders(X: np.ndarray, y: np.ndarray, batch_size: int = 128, validation_fraction: float = 0.15):
    X_tensor = torch.tensor(X, dtype=torch.long)
    y_tensor = torch.tensor(y, dtype=torch.long)
    dataset = TensorDataset(X_tensor, y_tensor)
    val_size = max(1, int(len(dataset) * validation_fraction))
    train_size = len(dataset) - val_size
    generator = torch.Generator().manual_seed(SEED)
    train_ds, val_ds = random_split(dataset, [train_size, val_size], generator=generator)
    return (
        DataLoader(train_ds, batch_size=batch_size, shuffle=True),
        DataLoader(val_ds, batch_size=batch_size),
    )

test_loader = DataLoader(TensorDataset(torch.tensor(X_test, dtype=torch.long), torch.tensor(y_test, dtype=torch.long)), batch_size=256)
````

````python
class RNNClassifier(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_units, num_classes, rnn_type="gru", dropout_rate=0.35, embedding_weights=None, freeze_embeddings=False):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        if embedding_weights is not None:
            self.embedding.weight.data.copy_(torch.tensor(embedding_weights, dtype=torch.float32))
            self.embedding.weight.requires_grad = not freeze_embeddings
        rnn_cls = nn.GRU if rnn_type.lower() == "gru" else nn.LSTM
        self.rnn = rnn_cls(
            input_size=embedding_dim,
            hidden_size=hidden_units,
            batch_first=True,
            bidirectional=True,
            dropout=0.0,
        )
        self.dropout = nn.Dropout(dropout_rate)
        self.fc1 = nn.Linear(hidden_units * 2, 64)
        self.fc2 = nn.Linear(64, num_classes)
        self.activation = nn.ReLU()

    def forward(self, tokens):
        embedded = self.embedding(tokens)
        output, hidden = self.rnn(embedded)
        if isinstance(hidden, tuple):
            hidden = hidden[0]
        forward_last = hidden[-2]
        backward_last = hidden[-1]
        features = torch.cat([forward_last, backward_last], dim=1)
        features = self.dropout(self.activation(self.fc1(features)))
        return self.fc2(features)
````

````python
def train_model(model, train_loader, val_loader, epochs=6, learning_rate=1e-3):
    model = model.to(DEVICE)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    history = []
    best_state = None
    best_val_loss = float("inf")
    patience = 2
    stale_epochs = 0

    for epoch in range(1, epochs + 1):
        model.train()
        train_losses = []
        train_correct = 0
        train_total = 0
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(DEVICE), batch_y.to(DEVICE)
            optimizer.zero_grad()
            logits = model(batch_X)
            loss = criterion(logits, batch_y)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())
            train_correct += (logits.argmax(dim=1) == batch_y).sum().item()
            train_total += batch_y.size(0)

        model.eval()
        val_losses = []
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(DEVICE), batch_y.to(DEVICE)
                logits = model(batch_X)
                loss = criterion(logits, batch_y)
                val_losses.append(loss.item())
                val_correct += (logits.argmax(dim=1) == batch_y).sum().item()
                val_total += batch_y.size(0)

        row = {
            "epoch": epoch,
            "loss": float(np.mean(train_losses)),
            "accuracy": train_correct / train_total,
            "val_loss": float(np.mean(val_losses)),
            "val_accuracy": val_correct / val_total,
        }
        history.append(row)
        print(row)

        if row["val_loss"] < best_val_loss:
            best_val_loss = row["val_loss"]
            best_state = {key: value.cpu().clone() for key, value in model.state_dict().items()}
            stale_epochs = 0
        else:
            stale_epochs += 1
            if stale_epochs >= patience:
                break

    if best_state is not None:
        model.load_state_dict(best_state)
    return model, pd.DataFrame(history)

def predict_model(model, loader):
    model.eval()
    preds = []
    labels = []
    with torch.no_grad():
        for batch_X, batch_y in loader:
            logits = model(batch_X.to(DEVICE))
            preds.extend(logits.argmax(dim=1).cpu().numpy())
            labels.extend(batch_y.numpy())
    return np.array(labels), np.array(preds)

def evaluate_model(model, loader, model_name: str) -> dict:
    y_true, y_pred = predict_model(model, loader)
    metrics = {
        "model": model_name,
        "accuracy": accuracy_score(y_true, y_pred),
        "macro_precision": precision_score(y_true, y_pred, average="macro", zero_division=0),
        "macro_recall": recall_score(y_true, y_pred, average="macro", zero_division=0),
        "macro_f1": f1_score(y_true, y_pred, average="macro", zero_division=0),
        "weighted_f1": f1_score(y_true, y_pred, average="weighted", zero_division=0),
    }
    print(f"Evaluation for {model_name}")
    print(classification_report(y_true, y_pred, target_names=class_names, zero_division=0))
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", xticklabels=class_names, yticklabels=class_names, cmap="Blues")
    plt.title(f"Confusion matrix: {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.xticks(rotation=35, ha="right")
    plt.tight_layout()
    plt.show()
    return metrics

def plot_history(history: pd.DataFrame, title: str):
    display(history)
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    history[["loss", "val_loss"]].plot(ax=axes[0])
    axes[0].set_title(f"{title} loss")
    history[["accuracy", "val_accuracy"]].plot(ax=axes[1])
    axes[1].set_title(f"{title} accuracy")
    plt.tight_layout()
    plt.show()
````

## Model 1: Custom RNN Architecture

This model uses trainable embeddings initialized from scratch and a bidirectional recurrent layer.

Experiment comments:

- GRU is usually faster than LSTM and works well on short texts.
- LSTM has more gates and can sometimes model longer dependencies better.
- Larger hidden sizes increase capacity but may overfit noisy tweets.
- Dropout regularizes the classifier.

````python
train_loader, val_loader = make_loaders(X_train, y_train, batch_size=128)

custom_model = RNNClassifier(
    vocab_size=vocab_size,
    embedding_dim=128,
    hidden_units=64,
    num_classes=num_classes,
    rnn_type="gru",
    dropout_rate=0.35,
)

custom_model, custom_history = train_model(custom_model, train_loader, val_loader, epochs=6, learning_rate=1e-3)
plot_history(custom_history, "Custom BiGRU")
custom_metrics = evaluate_model(custom_model, test_loader, "Custom BiGRU")
custom_metrics
````

### Hyperparameter Experiments

Set `RUN_HYPERPARAMETER_EXPERIMENTS = True` to compare GRU vs LSTM and hidden-size changes. The default is `False` so the notebook remains fast for ordinary execution.

````python
RUN_HYPERPARAMETER_EXPERIMENTS = False

experiment_configs = [
    {"rnn_type": "gru", "hidden_units": 64, "dropout_rate": 0.35},
    {"rnn_type": "lstm", "hidden_units": 64, "dropout_rate": 0.35},
    {"rnn_type": "gru", "hidden_units": 128, "dropout_rate": 0.45},
]

experiment_results = []
if RUN_HYPERPARAMETER_EXPERIMENTS:
    for config in experiment_configs:
        model = RNNClassifier(vocab_size, 128, config["hidden_units"], num_classes, config["rnn_type"], config["dropout_rate"])
        model, _ = train_model(model, train_loader, val_loader, epochs=3, learning_rate=1e-3)
        name = f"Custom {config['rnn_type'].upper()} hidden={config['hidden_units']} dropout={config['dropout_rate']}"
        result = evaluate_model(model, test_loader, name)
        result.update(config)
        experiment_results.append(result)
    display(pd.DataFrame(experiment_results).sort_values("macro_f1", ascending=False))
else:
    print("Set RUN_HYPERPARAMETER_EXPERIMENTS = True to run the comparison grid.")
````

## Model 2: Transfer Learning with GloVe Embeddings

This model initializes the embedding matrix from pre-trained GloVe word vectors. The embedding layer remains trainable, so the vectors can adapt to pandemic-specific vocabulary.

This is transfer learning because the model starts with semantic information learned from a large external corpus rather than learning all word representations only from the COVID tweet dataset.

````python
GLOVE_PATH = Path("data/glove/glove.6B.100d.txt")
GLOVE_DIM = 100

def load_glove_embeddings(path: Path, embedding_dim: int = 100) -> dict[str, np.ndarray]:
    require_file(path)
    embeddings_index = {}
    with path.open("r", encoding="utf-8") as handle:
        for line in handle:
            values = line.rstrip().split(" ")
            word = values[0]
            vector = np.asarray(values[1:], dtype="float32")
            if vector.shape[0] == embedding_dim:
                embeddings_index[word] = vector
    return embeddings_index

def build_embedding_matrix(vocab: dict[str, int], embeddings_index: dict[str, np.ndarray], embedding_dim: int) -> np.ndarray:
    matrix = np.random.normal(0, 0.05, size=(len(vocab), embedding_dim)).astype("float32")
    matrix[vocab[PAD_TOKEN]] = np.zeros(embedding_dim, dtype="float32")
    hits = 0
    for word, idx in vocab.items():
        vector = embeddings_index.get(word)
        if vector is not None:
            matrix[idx] = vector
            hits += 1
    print(f"Embedding coverage: {hits} matched tokens ({hits / max(1, len(vocab)):.2%})")
    return matrix

glove_embeddings = load_glove_embeddings(GLOVE_PATH, GLOVE_DIM)
embedding_matrix = build_embedding_matrix(vocab, glove_embeddings, GLOVE_DIM)
````

````python
glove_model = RNNClassifier(
    vocab_size=vocab_size,
    embedding_dim=GLOVE_DIM,
    hidden_units=64,
    num_classes=num_classes,
    rnn_type="gru",
    dropout_rate=0.35,
    embedding_weights=embedding_matrix,
    freeze_embeddings=False,
)

glove_model, glove_history = train_model(glove_model, train_loader, val_loader, epochs=6, learning_rate=7e-4)
plot_history(glove_history, "GloVe BiGRU")
glove_metrics = evaluate_model(glove_model, test_loader, "GloVe BiGRU")
glove_metrics
````

## Model Comparison

````python
comparison_df = pd.DataFrame([custom_metrics, glove_metrics]).sort_values("macro_f1", ascending=False)
display(comparison_df)

ax = comparison_df.set_index("model")[["accuracy", "macro_f1", "weighted_f1"]].plot(kind="bar", figsize=(9, 5))
ax.set_title("Model comparison")
ax.set_ylabel("Score")
ax.set_ylim(0, 1)
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
````

## Final Summary and Most Interesting Insights

1. **Macro F1 is the most important score.** It treats all five sentiment classes equally, which matters because the class distribution is not perfectly balanced.

2. **Preprocessing has high impact.** Removing URLs and mentions reduces noise, while preserving hashtag words keeps useful pandemic and sentiment cues.

3. **The custom RNN learns task-specific vocabulary.** It can adapt to COVID-specific words but starts without outside semantic knowledge.

4. **GloVe transfer learning can improve stability.** Pre-trained embeddings provide useful word relationships before fine-tuning.

5. **Extreme sentiment classes are likely hardest.** `Extremely Positive` and `Extremely Negative` are close to their non-extreme neighbors, so many errors are expected between adjacent sentiment intensities.

6. **Future improvement:** a tweet-specific transformer such as BERTweet or RoBERTa would likely outperform the RNN models, but the RNN comparison is clearer for this assignment.
